```python
df = pd.read_csv(r"C:\Users\USER\Desktop\Datasets\worldpop1.csv")
df.head(20)
```

```python
df2= df.drop(['Country Code','Indicator Name', 'Indicator Code'], axis=1)
```

```python
df2.head()
```

```python
mycountry= 'Zimbabwe'
```

```python
country_data = df2[df2['Country Name']== mycountry]
```

```python
print(country_data)
```

```python
years= country_data.columns[2:].astype(int)
```

```python
population = country_data.iloc[:, 2:].values.flatten()
```

```python
plt.figure(figsize=(5,3))
plt.bar(years, population,color="green")
plt.xlabel('year')
plt.ylabel('population')
plt.title(f'Population of {mycountry} over the years')
plt.show()
```

```python
plt.hist(population, bins=20, color= "green", edgecolor ="black")
plt.show()
```
